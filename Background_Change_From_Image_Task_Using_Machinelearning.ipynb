# ***Import Some Libraries For Fatching Data Using API***

```python
import requests

from rembg import  remove

from PIL import Image

import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow

from io import BytesIO
```

# ***Create To Directories Original for Image In which Original Image With Background Is Present and Second is Mask In Which Background is Remove***

```python
import os


os.makedirs('original',exist_ok=True)
os.makedirs('masked',exist_ok=True)
```

# ***Taking The URL Of Image From Google***

```python
img_url = "https://images.pexels.com/photos/3473516/pexels-photo-3473516.jpeg?cs=srgb&dl=pexels-elijah-o%27donnell-3473516.jpg&fm=jpg"
img_name = img_url.split('/')[-1]
img_name
```

# ***Original Image Plot***

```python
img = Image.open(BytesIO(requests.get(img_url).content))
#img.save('original/'+img_name,format='jpeg')
plt.imshow(img)
```

# ***Path Of Output Image in Masked Directorie To Check The Background Is Remove Or Not***

```python
output_path = 'masked/'+img_name
output_path
```

# ***Code For Removing BackGround from Image***

```python
with open(output_path , 'wb') as f:
  input = open('original/'+img_name,'rb').read()
  subject=remove(input,alpha_matting=True)
  f.write(subject)

```

# ***Ploting Image After Removing Background***

```python
img = Image.open("/content/masked/pexels-photo-3473516.jpeg?cs=srgb&dl=pexels-elijah-o%27donnell-3473516.jpg&fm=jpg")

plt.imshow(img)
```

# ***Code For Set New BackGround In Image By Using Bytes Input and Output set height and weidth of Background Image As Per Image Size***

```python
background_img = "https://t4.ftcdn.net/jpg/05/52/73/41/360_F_552734101_iyXcmlRsctwQpp3T4VTzV5iYhd63vMfE.jpg"

background_img = Image.open(BytesIO(requests.get(background_img).content))

background_img = background_img.resize((img.width,img.height))

InsertbackgroundImg = Image.open(output_path)

background_img.paste(InsertbackgroundImg,(0,0),InsertbackgroundImg)

background_img.save('masked/background.jpg',format='jpeg')

```

# ***Ploting New Image After Set New Background in Image***

```python
AfterInsertBackgroundImage = Image.open("/content/masked/background.jpg")
plt.imshow(AfterInsertBackgroundImage)
```
